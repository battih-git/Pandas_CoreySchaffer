# Creating a DataFrane

```python
# Creating a data frame from a dictionary
people = {
    'first' : ['John', 'Jane', 'Huzefa'],
    'last' : ['Doe','Doe', 'Battiwala'],
    'email' : ['john_doe@gmail.com', 'jane_doe@gmail.com', 'huzefa_battiwala@gmail.com']
}

sample_df = pd.DataFrame(people)
sample_df
```

```python
type(sample_df) # This is the data frame which has rows and columns
```

```python
sample_df['email']
```

# What is a Series

```python
type(sample_df['email']) # This is the series which is one dimensional and only have rows. It has the indexes similar to data frame
```

```python
# we can access the email column in similar way with below syntax
sample_df.email
# Most preferable way to accessing the column is through square brackets. Let's say if there is a column name and if we call df.count then it would return the count instead of column values.
```

# Accessing multiple columns of data frame

```python
# To access specific columns more than 1 then we can pass it in a list
sample_df[['first','last']]
```

```python
columns_list = ['first','last']
sample_df[columns_list]
```

```python
# Access a column
sample_df.columns
```

```python
# Access rows: There are two ways loc and iloc
# iloc
sample_df.iloc[0]
```

```python
# Accessing multiple rows with iloc
sample_df.iloc[[0,1]]
```

```python
# Accessing rows and column with iloc
sample_df.iloc[[0,1],2]
```

```python
#loc. Labels are used to access the value. Currently the index value for the row is number
sample_df.loc[[0,1],['email']]
```

```python
sample_df.loc[0:1,['email']]
```
