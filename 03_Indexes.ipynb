```python
import pandas as pd
# Creating a data frame from a dictionary
people = {
    'first' : ['John', 'Jane', 'Huzefa'],
    'last' : ['Doe','Doe', 'Battiwala'],
    'email' : ['john_doe@gmail.com', 'jane_doe@gmail.com', 'huzefa_battiwala@gmail.com']
}

sample_df = pd.DataFrame(people)
sample_df
```

```python
# We can assign index which is unique
sample_df.set_index('email',inplace=True)
```

```python
sample_df.index
```

```python
# Accessing the rows of df with specific index values, instead of range values.
sample_df.loc['john_doe@gmail.com']
```

```python
# Now the index values are changed, we cannot use range values with loc
sample_df.loc[0] # This will give a keyerror
```

```python
# If we want to reset the index. we can use reset_index
sample_df.reset_index(inplace=True)
```
